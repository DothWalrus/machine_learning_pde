```python
# Imports
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
import tikzplotlib

# Load training data
x_train = -50 + np.random.random((25000,1))*100
y = np.sin(x_train) #+ np.random.normal((25000))*.1
# Define model
model = Sequential()
model.add(Dense(100, input_dim=1, activation='tanh'))
model.add(Dense(100, activation='tanh'))
model.add(Dense(100, activation='tanh'))
model.add(Dense(100, activation='tanh'))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam')
model.fit(x_train, y, epochs=20, batch_size=50)
```

```python
x = np.sort(-50 + np.random.random((2500,1))*100, axis=None)
predictions = model.predict(x)
print(np.shape(y))
#plot
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, np.sin(x), color='blue', label='u(x)=sin(x)')
ax.plot(x, predictions,'r--', label='g(x)')
#ax.legend()

tikzplotlib.save("learn_sine.tex")
```
